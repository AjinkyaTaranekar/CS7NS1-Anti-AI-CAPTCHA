```python
import pandas as pd
import json
```

```python
df = pd.read_csv("./data/train-data.csv", sep=";", quotechar="'")
    
# Load symbols map
symbols = pd.read_csv("./data/symbols.csv", sep=";", quotechar="'")
symbol_map = {row['symbol_id']: row['latex'] for _, row in symbols.iterrows()}

# Map and filter
df['char'] = df['symbol_id'].map(symbol_map)
df    
```

```python
unique = df.drop_duplicates(subset='char', keep='first')[['char', 'data']].copy()

unique
```

```python
symbols = {}
for _, row in unique.iterrows():
    symbols[row['char']] = json.loads(row['data'])
symbols
```

```python
import json

with open("symbol_mouse_data.json", "w") as f:
    # Orient the JSON for easier reading
    json.dump(symbols, f, indent=4)
```
